```python
# Crie duas listas , lista1 e lista2 , cada uma contendo pelo menos 4

lista11 = []
for _ in range(4):
    numero = int(input("Digite um número inteiro: "))
    lista11.append(numero)

lista22 = []
for _ in range(4):
    nome = str(input("Digite um nome: "))
    lista22.append(nome)

lista_concatenada = lista11 + lista22
print("Lista completa: ", lista_concatenada)
```
