```python
## ESCREVA UM PROGRAMA EM PYTHON QUE PEÇA PARA O USUARIO FAZER DUAS JOGADAS ENTRE PEDRA, PAPE E TESOURA
## UMA PARA O PRIMEIRO JOGADOR, OUTRA PARA O SEGUNDO.DEPOIS DIGA QUAL DELES GANHOU A DISPUTA

##LEMBRANMDO QUE
# ESCOLHAS IGUAIS = EMPATE;
# PEDRA GANHA DE TESOURA E PERDE PARA PAPEL;
# TESOURA GANHA DE PAPEL E PERDE PARA PEDRA;
# PAPEL GANHA DE PEDRA E PERDE PARA TESOURA.
```

```python
def jogar():
    jogar_novamente = input("Deseja jogar novamente? (S/N): ")
    if jogar_novamente.upper() == 'S':
        jogoptp()
    elif jogar_novamente.upper() == 'N':
        print("Obrigado por jogar!")
    else:
        print("Opção inválida.")
        jogar()

def jogoptp():
    jogar = input("Deseja jogar um jogo? (S/N): ")
    if jogar.upper() == 'S':
        print("Então vamos jogar Pedra, Papel e Tesoura!")
        jogo1()
    elif jogar.upper() == 'N':
        print("Tudo bem, até a próxima!")
    else:
        print("Opção inválida.")
        jogoptp()

def jogo1():
    print("Primeiro, digite seus nomes:")
    nome1 = input("Nome do Primeiro Jogador: ")
    nome2 = input("Nome do Segundo Jogador: ")
    print("Ok, agora que temos os nomes, vamos jogar!")

    jogador1 = input(f"{nome1}, escolha entre pedra, papel ou tesoura: ")
    jogador2 = input(f"{nome2}, escolha entre pedra, papel ou tesoura: ")

    print("Vamos ver quem ganhou!")

    if jogador1 == jogador2:
        print("Empate!")
    elif jogador1 == "pedra":
        if jogador2 == "tesoura":
            print(f"{nome1} venceu!")
        else:
            print(f"{nome2} venceu!")
    elif jogador1 == "papel":
        if jogador2 == "pedra":
            print(f"{nome1} venceu!")
        else:
            print(f"{nome2} venceu!")
    elif jogador1 == "tesoura":
        if jogador2 == "papel":
            print(f"{nome1} venceu!")
        else:
            print(f"{nome2} venceu!")
    else:
        print("Opção inválida, por favor escolha entre pedra, papel ou tesoura.")

    jogar()

jogoptp()
```
